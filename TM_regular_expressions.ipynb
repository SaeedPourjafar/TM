### Text mining practices w/ regular expressions

##### 1. Exercise ----

```python
# Vector of strings is given

vector = ["emoticon", ":)", "symbol", "$^$"]
vector
```

```python
# a) string of 3 characters with the letter o in the middle
import re
re.findall(".[o].", ' '.join(vector))
```

```python
# b) expression "emoticon"
re.search("emoticon", ' '.join(vector)).group()
```

```python
# c) expression ":)"
re.search(":\)", ''.join(vector)).group()
```

```python
# d) expression "$^$"
re.search("\$\^\$", ' '.join(vector)).group()
```

##### 2. Exercise ----

```python
# Fist we should start off by downloading the brown corpus from nltk package
import nltk
nltk.download('brown')
```

```python
# Let's examine the categories and their repective length in brown corpus
from nltk.corpus import brown
for i in brown.categories():
    print(i+':' ,len(", ".join(brown.words(categories=i))))
```

```python
# And continue the exercise with science_fiction category
sf = brown.words(categories='science_fiction')
sf
```

```python
# a) all words containing the expression "yes" (add the parameter match=T)
# Please note that although there were instances of "Yes" but they were all located in the beginning of the sentence.
# The question asked for "yes" -and not "Yes"- therefore only instances of "eyes" are shown
re.findall(r'\w*yes\w*', ' '.join(sf))
```

```python
# b) all words starting with "w"
re.findall(r'\b[w].*?\b', ' '.join(sf))
```

```python
# c) all words ending with "x"
re.findall(r'\w*x\b', ' '.join(sf))
```

##### 3. Exercise ----

```python
# a) all words starting with a vowel
re.findall(r"\b[(aeiou|AEIOU)].*?\b", ' '.join(sf))
```

```python
# b) all words that start only with a consonant
a = re.findall(r"\b[(b-df-hj-np-tv-z|B-DF-HJ-NP-TV-Z)].*?\b", ' '.join(sf))
a
```

```python
# b) all words that start only with a consonant
# Another way with exclusion syntax to get the same result:
b = re.findall(r"\b[^\W_aeiouAEIOU0-9].*?\b", ' '.join(sf))
a == b
```

```python
# c) # all words ending with "ing" or "ise"
re.findall(r"\w*ing|\w*ise\b", ' '.join(sf))
```

```python
# d) all words ending with "ed" but not with "eed"
re.findall(r"\w*.[a-df-z]ed\b", ' '.join(sf))
```
